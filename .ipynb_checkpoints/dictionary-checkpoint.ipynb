```python
from os import listdir#讀檔名
from os.path import isfile, join#檔名路徑
from bs4 import BeautifulSoup as bs#解析檔案
import csv#讀取CSV檔案

all_file_path = "C:\\Users\\BigData\\wordlist"#字典存放路徑
files_list = [ f for f in listdir(all_file_path) if isfile(join(all_file_path,f))]#路徑中檔案名稱
f = file('C:\\Users\\BigData\\output.txt', 'a') # open for 'w'riting
#建立對照檔字典
compare_dictionary={1:"dictionary_name"}#建立空字典
compare = open('C:\\Users\\BigData\\compare.csv', 'r')#開啟對照檔
for row in csv.DictReader(compare,["name", "number"]):#將對照檔中每一個row讀入
    str_number=int(row["number"].decode('big5'))#將對照檔之數字轉成int方便建立字典
    compare_dictionary[str_number]=(row["name"].decode('big5'))#將對照檔做成compare_dictionary字典(key為檔名中的數字value為字典名稱)
compare.close()#將compare檔關閉 
#開始開檔順序印出
for number_of_file in range(100,194):#測試用小數字代替#len(files_list)#順序將檔案取出
    one_file_name = files_list[number_of_file]#將當次的檔案其檔名取出
    file_name_str = str(one_file_name)#將檔名建成字串方便轉乘路徑
    whole_file_path = 'C:\\Users\\BigData\\wordlist\\'+ file_name_str#建立完整路徑以便讀取該檔案
    one_file_content = open(whole_file_path,'r')#將該檔案讀取出來
    soup = bs(one_file_content)#將該檔案用soup解析(此為特殊用法不適合一般excel)
    
    #建立迴圈之前須將對應之字典名稱取出已變最後輸出成另一欄位
    split_list=file_name_str.split('_')#將檔名拆開
    middle_number=int(split_list[1])#取出中間數字當key  特別注意取出後須轉成int方能呼叫value
    
    #print compare_dictionary[middle_number]  #呼叫該字典名稱
    
    #開始跑回圈輸出並加入對照檔之字典名稱
    for row in soup.select('tr'):#取出其中一列
        
        #print row.select('td')[2].text.encode('utf-8') + ','+ compare_dictionary[middle_number].encode('utf-8')  #取出中文字詞並加入字典名稱
        line = row.select('td')[2].text.encode('utf-8') +','+ compare_dictionary[middle_number].encode('utf-8')
        f.write(line + '\n')
        
    print number_of_file
f.close()      
one_file_content.close()#跑完全部檔案後關閉檔案
```
